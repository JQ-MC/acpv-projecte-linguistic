```python
import pandas as pd
import numpy as np
import re
```

```python
dataset = pd.read_csv("data.csv")
```

```python
dataset.head()
```

```python
dataset["En cas d'altres, quina?"].loc[dataset["En cas d'altres, quina?"].notnull()].str.lower().str.contains("fall")
```

```python
dataset["Categoria"].loc[dataset["En cas d'altres, quina?"].notnull()].loc[dataset["En cas d'altres, quina?"].loc[dataset["En cas d'altres, quina?"].notnull()].str.lower().str.contains("fall")] = "Falles"
```

```python
dataset['Data'] = pd.to_datetime(dataset['Data'], format='%d/%m/%Y', errors="coerce")
dataset['Marca de temps'] = pd.to_datetime(dataset['Marca de temps'], format='%d/%m/%Y %H:%M:%S')
```

```python
dataset["PreuMitja"] = dataset.Preu.str.findall(r"\d+[,'.]?\d*")
```

```python
dataset = dataset[["Preu","PreuMitja"]]
```

```python
dataset["PreuMitja"].loc[dataset.Preu.notnull()] = dataset.loc[dataset.Preu.notnull()].PreuMitja.apply(lambda x: np.mean([float(num.replace("'", ".").replace(",", ".")) for num in x]))
```
